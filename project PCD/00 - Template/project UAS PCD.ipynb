```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('godong.jfif')
img_resize = cv2.resize(img, (208, 278))

edged = cv2.Canny(img_resize, 150, 260) 

cv2.imshow("Original Image",img_resize)
cv2.imshow("Edged Image", edged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
